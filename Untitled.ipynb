Bank Returns defined as:
  =(p(t) - p(t-1))/p(t-1)
  =(p(t)/p(t-1)) - 1
 Here p(t) -> closing price of bank stocks at time t

Creating new section of bank returns by each banks

```python
okk = pd.DataFrame()
```

```python
for tick in tickers:
    okk[tick+' Return'] = df[tick]['Close'].pct_change()
```

```python
okk #gives returns dataframe
```

Visualing each bank's return and comparing their return with other bank ones

```python
sns.pairplot(okk[1:])  #in the below plotting almost every plot look normal except for C returns due to November 2008 collapse 
                       #due to financial crisis
```
